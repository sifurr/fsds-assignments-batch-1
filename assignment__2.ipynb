#### 1. Email Validator (String Filtering + Loops)

ðŸ’¼ Real-world use: Used in signup forms or data cleaning tasks.

Task:
Check if a given email is valid â€” must contain â€œ@â€ and end with â€œ.comâ€ or â€œ.orgâ€.

Example:

Input: hello@inceptionbd.com

Output: Valid Email

```python
user_input = input("Enter your Email Address: ", )
email_address = user_input.strip().lower()

at_sign_flag = False
is_valid = False

for char in email_address:
    if len(email_address) > 5:
        if char == "@":
            at_sign_flag = True
            break

if not at_sign_flag:
    print("Not valid")
else:
    if len(email_address) >= 4 and email_address[-4:] == ".com":
        is_valid = True  
    elif len(email_address) >= 4 and email_address[-4:] == ".org":
        is_valid = True

    if is_valid:
        print("Valid Email")
    else:
        print("Not valid")
```

#### 2. Count Word Frequency in Customer Reviews

ðŸ’¼ Real-world use: Analyzing reviews or comments in text analytics.

Task:
Given a review text, count how many times each word appears (ignore case).

Example:
Input: "This product is good and this service is excellent"

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
user_input = input("Enter a customer review: ")
customer_review = user_input.lower().strip().replace(",", "").replace(".", "").split(" ")

word_list = []
frequency_count = []
counter_2 = 0

while counter_2 < len(customer_review):
    word = customer_review[counter_2]

    if word not in word_list:
        word_list.append(word)
        frequency_count.append(1)
    else:
        index = word_list.index(word)
        frequency_count[index] += 1

    counter_2 += 1

print("{", end="")
for i in range(len(word_list)):
    if i != len(word_list) - 1:
        print(f"'{word_list[i]}': {frequency_count[i]}", end=", ")
    else:   
        print(f"'{word_list[i]}': {frequency_count[i]}", end="")
print("}")
```

#### 3. Password Strength Checker
ðŸ’¼ Real-world use: Used in user authentication systems.

Task:
Check if a password is strong â€” must have at least one uppercase, one lowercase, one digit, and one
special character.

Example:

Input: "Hello@123"

Output: Strong Password

```python
import string

password = input("Enter your password: ")
password_list = list(password.strip())

password_upper_case = len([char for char in password_list if char.isupper()])
password_lower_case = len([char for char in password_list if char.islower()])
password_digit = len([char for char in password_list if char.isdigit()])
password_spec_char = 0

special_characters = string.punctuation
for char in password_list:
    if char in special_characters:
        password_spec_char += 1

if (password_upper_case > 0 and password_lower_case > 0 and 
    password_digit > 0 and password_spec_char > 0):
    print("Strong Password")
else:
    print("Password must have at least one uppercase, one lowercase, one digit, and one special character")
```

#### 6. Remove Duplicates from Contact List
ðŸ’¼ Real-world use: Data cleaning in CRM or user databases.

Task:

Given a list of names (comma separated), remove duplicates.

Example:

Input: "Rahim, Karim, Rahim, Sultana"

Output: "Rahim, Karim, Sultana"

```python
user_input = input("Enter names: ")
names = user_input.strip().split(",")

unique_names = []

counter_1 = 0 
while counter_1 < len(names):
    name = names[counter_1].strip().lower().title()
    if name not in unique_names: 
        unique_names.append(name)
    counter_1 += 1

counter_2 = 0
print('"', end="")
while counter_2 < len(unique_names):
    if counter_2 != len(unique_names) - 1:
        print(unique_names[counter_2], end=", ")
    else:
        print(unique_names[counter_2], end="")
    counter_2 += 1
print('"')
```

#### 7. Detect Suspicious Words in Text (Moderation)

ðŸ’¼ Real-world use: Used in chat moderation or email filtering.

Task:

If any â€œbannedâ€ words like ['spam', 'scam', 'fake'] are found in a text, print â€œWarningâ€.

Example:

Input: "This is a scam offer"

Output: Warning: Suspicious content detected

```python
user_input = input("Enter a sentence: ")
sentence = user_input.strip().lower()

banned_words = ["spam", "scam", "fake"]

counter = 0
flag = False
while counter < len(banned_words):
    word = banned_words[counter]

    if sentence.find(word) != -1:
        flag = True
    
    counter += 1

if flag:
    print("Warning: Suspicious content detected")
else:
    print("No suspicious content found")
```

#### 9. Count Hashtags and Mentions (Social Media Analytics)

ðŸ’¼ Real-world use: Used in analyzing social media posts.

Task:

Count how many hashtags (#) and mentions (@) appear in a post.

Example:

Input: "Check out #AIWorkshop and follow @inceptionbd"

Output: Hashtags = 1, Mentions = 1

```python
user_input = input("Write your post: ")
words = user_input.strip().replace(",", "").replace(".", "").split(" ")

hash_counter = 0
mention_counter = 0
counter = 0

while counter < len(words):
    word = words[counter]

    if word.startswith("#"):
        hash_counter += 1
    if word.startswith("@"):
        mention_counter += 1

    counter += 1

print(f"Hashtags = {hash_counter}, Mentions = {mention_counter}")
```

#### 10. Word Reversal for Data Encryption
ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding.

Task:

Reverse each word in a sentence while keeping word order the same.

Example:

Input: "AI is fun"

Output: "IA si nuf"

```python
user_input = input("Enter your sentence: ")
words = user_input.strip().split(" ")
reversed_words = [word[::-1] for word in words] 
print(' '.join(reversed_words))
```
